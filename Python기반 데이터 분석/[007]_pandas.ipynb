# III. pandas - Visualization

* https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html

>## 1) 선 그래프

* <font color = 'blue'>kind = 'line'</font>
* style: '-','--','-.', 's-', 'o-', '^-'
* linewidth: 숫자
* color: 16진수 RGB
* figsize: inch

```
DF1[['Height']].plot(kind = 'line',
                     style = '--',
                     linewidth = 2,
                     color = '#FF0000',
                     grid = True,
                     figsize = (10, 7))
```

>## 2) 막대 그래프

* <font color = 'blue'>kind = 'bar'</font>
* width: 0~1
* rot: 0~360 (범주 이름 rotation )
* fontsize: 글자 크기

```
DF1[['Height', 'Weight']].plot(kind = 'bar',
                               width = 0.5,
                               rot = 45,
                               fontsize = 15,
                               figsize = (10, 7))
```

* <font color = 'blue'>kind = 'barh' </font>    (bar-horizontal)

```
DF1[['Height', 'Weight']].plot(kind = 'barh',
                               rot = 5,
                               color = ['#FA5800', '#A20025'],
                               fontsize = 12,
                               figsize = (7, 10))
```

>## 3) 히스토그램

* <font color = 'blue'>kind = 'hist'</font>
* bins: 구간 개수
* alpha: 0~1 [투명도: 0(투명), 1(불투명)]

```
DF1['Height'].plot(kind = 'hist',
                   bins = 5,
                   alpha = 0.5,
                   figsize = (10, 7))
```

```
DF1['Height'].plot(kind = 'hist',
                   bins = [157, 160, 170, 175, 180, 182],
                   alpha = 0.5,
                   figsize = (10, 7))
```

>## 4) 상자 그래프

* <font color = 'blue'>kind = 'box'

```
DF1['Height'].plot(kind = 'box',
                   figsize = (7, 9))
```

>## 5) 산점도 - 2가지 연속형 변수의 상관관계 확인

* <font color = 'blue'>kind = 'scatter'</font>
* s: 숫자
* x, y 두가지 변수 필요

```
DF1[['Height', 'Weight']].plot(kind = 'scatter',
                               x = 'Height',
                               y = 'Weight',
                               s = 50,
                               figsize = (10, 7))
```

>## 6) 파이 그래프

* <font color = 'blue'>kind = 'pie'</font>
* 명목형 변수는 숫자형으로 변환 (ex. 카운트)

```
DF1.BloodType.value_counts().plot(kind = 'pie',
                                  autopct = '%.1f%%',
                                  fontsize = 15,
                                  figsize = (10, 10))
```
